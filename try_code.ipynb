```python
import pandas as pd

def print_problematic_rows(file_path, start_row=1560, num_rows=20):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        for i in range(start_row, start_row + num_rows):
            if i < len(lines):
                print(f"Row {i + 1}: {lines[i]}")
            else:
                print(f"Row {i + 1}: [No data]")

path = "Datasets/movies.csv"
print_problematic_rows(path)
```
